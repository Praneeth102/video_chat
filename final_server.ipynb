```python
import cv2
import socket
import time
import base64
import pickle
import threading#server code
```

```python
p1=socket.socket()
p1.setsockopt(socket.SOL_SOCKET ,socket.SO_REUSEADDR,1)
p1_port=1485
ip=""
p1.bind((ip,p1_port))
p1.listen()
p1_s,p1_addr=p1.accept()
print(p1_addr)
```

```python
def recvimage():
    i=0
    while True:
        time.sleep(0.2)
        try:
            data=p1_s.recv(10000000)   
            try:
                ima=pickle.loads(data)
                image=cv2.imdecode(ima,cv2.IMREAD_COLOR)
                if image is not None:
                    #print(image)
                    print(image.shape)
                    cv2.imshow('2nd_per',image)
                    if cv2.waitKey(100) == 13:
                        break
            
            except:
                pass
        except:
            pass
    cv2.destroyAllWindows()
```

```python
def sendimage():
    cap=cv2.VideoCapture(0)
    while True:
        time.sleep(0.2)
        ret,pic=cap.read()
        #print(pic.shape)
        photo=pic.reshape(480,640,3)
        ret,buffer=cv2.imencode('.jpg',photo)
        encodimage=pickle.dumps(buffer)
        p1_s.send(encodimage)
```

```python
recev=threading.Thread(target=recvimage)
sendd=threading.Thread(target=sendimage)
```

```python
recev.start()
sendd.start()
```
